```python
# Import modules
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Read in the Data

# Create file path
file = Path('../datasets/cleaned_gbp_dataset.csv')

# Read in Data
gbp_df = pd.read_csv(file, encoding="UTF-8", low_memory=False)
```

```python
gbp_df.head()
```

```python
gbp_df['location'].unique().tolist()
```

```python
# Convert 'year' index to integer type
columns_to_plot = gbp_df.columns.difference(['location', 'year'])
gbp_df.loc[:, 'year'] = gbp_df['year'].astype(int)

# Define a colormap
colors = plt.cm.tab20(np.linspace(0, 1, len(gbp_df['location'].unique())))

# Plotting loop for each column
for i, column in enumerate(columns_to_plot):
    plt.figure(figsize=(12, 8))
    
    for j, country in enumerate(gbp_df['location'].unique()):
        country_data = gbp_df[gbp_df['location'] == country]
        plt.plot(country_data['year'], country_data[column], label=country, color=colors[j])

    plt.title(f'{column.capitalize()} by Country')
    plt.xlabel('Year')
    plt.ylabel(f'{column.capitalize()}')
    
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    plt.tight_layout()
    
    plt.show()
```
