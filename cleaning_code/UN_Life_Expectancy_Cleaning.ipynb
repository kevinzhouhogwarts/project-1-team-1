```python
# Import modules
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create file path
life_expectancy = Path("../Resources/Life_Expectancy_UNData.csv")
# Read in Data
life_expectancy_df = pd.read_csv(life_expectancy, low_memory=False)

display(life_expectancy_df)
```

```python
# Clean the data

# Strip whitespace from column titles
life_expectancy_df.columns = life_expectancy_df.columns.str.strip()
# Rename useful columns as needed 
renamed_df = life_expectancy_df.rename(columns={"Region, subregion, country or area *":"Country", 
                                                "Life Expectancy at Birth, both sexes (years)":"Life Expectancy",
                                                "Total Population, as of 1 July (thousands)":"Population"
                                               })
renamed_df
```

```python
# Extract necessary columns

final_df = renamed_df[['Country', 'Year', 'Life Expectancy', 'Population', 'Type', 'Male Population, as of 1 July (thousands)', 'Female Population, as of 1 July (thousands)', 'Male Life Expectancy at Birth (years)', 'Female Life Expectancy at Birth (years)']]
final_df
```

```python
final_df.dtypes
```

```python
# Drop rows with null values in 'Life Expectancy' and 'Year'
final_df = final_df.dropna(subset=['Life Expectancy', 'Year'])

# Replace non-numeric values with NaN and convert columns to numeric
final_df.loc[:, 'Life Expectancy'] = pd.to_numeric(final_df['Life Expectancy'], errors='coerce')
final_df.loc[:, 'Male Life Expectancy at Birth (years)'] = pd.to_numeric(final_df['Male Life Expectancy at Birth (years)'], errors='coerce')
final_df.loc[:, 'Female Life Expectancy at Birth (years)'] = pd.to_numeric(final_df['Female Life Expectancy at Birth (years)'], errors='coerce')

# # Fill NaN values with 0 for Life Expectancy 
# final_df['Life Expectancy'] = final_df['Life Expectancy'].fillna(0)
# final_df['Male Life Expectancy at Birth (years)'] = final_df['Male Life Expectancy at Birth (years)'].fillna(0)
# final_df['Female Life Expectancy at Birth (years)'] = final_df['Female Life Expectancy at Birth (years)'].fillna(0)

# Modify data types as needed using .loc
final_df.loc[:, 'Year'] = final_df['Year'].astype(int)
final_df.loc[:, 'Life Expectancy'] = final_df['Life Expectancy'].astype(float)
final_df.loc[:, 'Male Life Expectancy at Birth (years)'] = final_df['Male Life Expectancy at Birth (years)'].astype(float)
final_df.loc[:, 'Female Life Expectancy at Birth (years)'] = final_df['Female Life Expectancy at Birth (years)'].astype(float)

# Display the modified DataFrame
final_df.head()
```

```python
# # Fill NaN values with 0 for Population columns
# final_df['Population'] = final_df['Population'].fillna(0)
# final_df['Male Population, as of 1 July (thousands)'] = final_df['Male Population, as of 1 July (thousands)'].fillna(0)
# final_df['Female Population, as of 1 July (thousands)'] = final_df['Female Population, as of 1 July (thousands)'].fillna(0)


# Convert columns to string, replace spaces, and convert to numeric, coercing errors to NaN
final_df.loc[:, 'Population'] = pd.to_numeric(final_df['Population'].astype(str).str.replace(' ', ''), errors='coerce')
final_df.loc[:, 'Male Population, as of 1 July (thousands)'] = pd.to_numeric(final_df['Male Population, as of 1 July (thousands)'].astype(str).str.replace(' ', ''), errors='coerce')
final_df.loc[:, 'Female Population, as of 1 July (thousands)'] = pd.to_numeric(final_df['Female Population, as of 1 July (thousands)'].astype(str).str.replace(' ', ''), errors='coerce')

# Display the cleaned DataFrame
final_df
```

```python
# Specify the file path where you want to save the CSV file
csv_file_path = '../output/csv/life_expectancy.csv'

# Write final_df to CSV
final_df.to_csv(csv_file_path, index=False)

print(f"DataFrame successfully written to {csv_file_path}")
```
