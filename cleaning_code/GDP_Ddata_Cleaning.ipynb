```python
# Import modules
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Read in the Data

# Create file path
gdp_file = Path('../Resources/GDP_Data_MPD.csv')

# Read in Data
#df = pd.read_csv(file, encoding="UTF-8", low_memory=False)
gdp_file_df = pd.read_csv(gdp_file)
display(gdp_file_df)
```

```python
# Clean the data

# Strip whitespace from column titles
gdp_file_df.columns = gdp_file_df.columns.str.strip()
# Rename useful columns as needed 
renamed_df = gdp_file_df.rename(columns={"country":"Country", 
                                                "year":"Year",
                                                "gdppc":"GDP per Capita",
                                                "region":"Region"
                                               })
renamed_df
```

```python
# Manual mapping dictionary to match names of few countries with UN life Expectancy file
country_mapping = {
    "United States": "United States of America",
    "D.P.R. of Korea": "Dem. People's Republic of Korea",
    "D.R. of the Congo": "Democratic Republic of the Congo",
    "U.R. of Tanzania: Mainland": "United Republic of Tanzania",
    "Taiwan, Province of China": "China, Taiwan Province of China",
    "TFYR of Macedonia": "North Macedonia",
    "Turkey": "TÃ¼rkiye"
}
# Standardize country names in gdp file
renamed_df['Country'] = renamed_df['Country'].replace(country_mapping)
renamed_df
```

```python
# Specify the file path where you want to save the CSV file
csv_file_path = '../output/csv/final_gdp.csv'

# Write final_df to CSV
renamed_df.to_csv(csv_file_path, index=False)

print(f"DataFrame successfully written to {csv_file_path}")
```
